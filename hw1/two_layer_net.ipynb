## Теоретический вопрос 1

Мы просили реализовать только ReLU, но в нейросетях можно использовать много разных функций активации, у каждой есть свои плюсы и минусы. Одна из частых проблем — исчезающие градиенты, когда во время обратного распространения градиент становится равным нулю или очень близким к нулю.

Какие из перечисленных ниже функций страдают от этой проблемы? Если рассматривать функции в одномерном случае, то для каких входных значений это происходит?

1. `sigmoid`
2. `ReLU`
3. `Leaky RELU`

### Ваш ответ: 

* Sigmoid Страдает ли: Да.При каких значениях: При больших по модулю значениях x (как больших положительных, так и больших отрицательных). В этих областях функция ее значения приближаются к 1 или к 0, и её производная стремится к нулю. Чем дальше значение x от нуля, тем меньше градиент.

* ReLU Страдает ли: Да, в специфической форме ("мертвые нейроны"). При каких значениях: При любых отрицательных значениях входа (x < 0). Для всех отрицательных чисел производная ReLU строго равна нулю. Если на протяжении обучения нейрон начинает получать только отрицательные значения, градиент через него перестает проходить полностью, и веса этого нейрона больше никогда не обновляются.
    
* Leaky ReLU Страдает ли: Задумывалась так, чтобы не страдать от этой проблемы. При каких значениях: Эта функция была специально создана для решения проблемы "мертвых нейронов". При положительных входах её градиент равен 1 (как у ReLU), а при отрицательных (x < 0) градиент равен не нулю, а небольшой константе. Таким образом, градиент никогда не становится равным нулю, что позволяет весам продолжать обновляться даже при отрицательных входах.

# Подбор гиперпараметров

**Что здесь не так?** По графикам выше видно, что функция потерь убывает почти линейно. Это часто говорит о том, что скорость обучения слишком мала. Кроме того, между обучающей и валидационной точностью почти нет разрыва, а значит, модель, скорее всего, имеет недостаточную ёмкость и её стоит увеличить.

С другой стороны, слишком большая модель может начать переобучаться: тогда обучающая точность станет заметно выше валидационной. Ваша цель — подобрать такие гиперпараметры, чтобы получить качество выше `48%` на валидации.
При выборе варьируемых гиперпараметров модно отталкиваться от параметров TwoLayerNet

````python
best_model = None

################################################################################
#                              YOUR CODE                                       #
################################################################################

best_model = None
best_val_acc = -1
results = {}

# Увеличиваем размер скрытого слоя
hidden_sizes = [100, 150] 
# learning rate чуть больше стандартного
learning_rates = [1e-3, 1.5e-3, 2e-3] 
# L2-регуляризация
regularization_strengths = [1e-4, 1e-3, 1e-2] 

for hs in hidden_sizes:
    for lr in learning_rates:
        for reg in regularization_strengths:
            print(f"Тренировка: hidden_size={hs}, lr={lr}, reg={reg}")
            
            model = TwoLayerNet(input_dim=input_size, 
                                hidden_dim=hs, 
                                num_classes=num_classes, 
                                reg=reg)
            
            solver = Solver(model, data,
                            update_rule='sgd',
                            optim_config={'learning_rate': lr},
                            lr_decay=0.95,
                            num_epochs=10, 
                            batch_size=100,
                            verbose=False) 
            
            solver.train()
            
            val_acc = solver.best_val_acc
            results[(hs, lr, reg)] = val_acc
            print(f"Точность на валидации: {val_acc}")
            
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                best_model = model

print(f"Лучшая точность на валидации: {best_val_acc}")
````
